```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
import matplotlib.pyplot as plt
import seaborn
import numpy as np

%matplotlib inline
```

```python
def f(x):
    return x * np.sin(x * 2 * np.pi) if x < 0 else -x * np.sin(x * np.pi) + np.exp(x / 2) - np.exp(0)

f = np.vectorize(f)
```

```python
x = np.linspace(-3, 3, 1000)
y = f(x)
```

```python
def baseline_model():
    model = Sequential()
    model.add(Dense(50, input_dim=1, activation='tanh', init='he_normal'))
    model.add(Dense(50, activation='tanh', init='he_normal'))
    model.add(Dense(1, activation='linear', init='he_normal'))
    sgd = SGD(lr=0.01, momentum=0.9, nesterov=True)
    model.compile(loss='mean_squared_error', optimizer=sgd)
    return model
```

```python
model = baseline_model()
model.fit(x, y, nb_epoch=400, verbose=False)
```

```python
plt.scatter(x, y, color='black', antialiased=True)
plt.plot(x, model.predict(x), color='magenta', linewidth=2, antialiased=True)
```
